```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

font = {'family': 'serif',
        'color':  'black',
        'weight': 'normal',
        'size': 16,
        }
```

```python
with open('results/sfd-no-poisoning/result_FEMNIST_balanced/col_performance.pkl', 'rb') as f:
    col_perf = pickle.load(f)

plt.figure(figsize=(8, 6))
for i in range(len(col_perf)):
    plt.plot(col_perf[i])

plt.title('FEMNIST IID', fontdict=font)
plt.xlabel('No. of Epoch', fontdict=font)
plt.ylabel('Test Accuracy', fontdict=font)
plt.xticks(list(range(21)))
plt.plot()
```

```python
with open('results/sfd-poison/result_FEMNIST_balanced/col_performance.pkl', 'rb') as f:
    col_perf = pickle.load(f)

accuracy_values = list(col_perf.values())

# Plotting the box plots
plt.boxplot(accuracy_values)

# Setting x-axis labels
plt.xticks(range(1, len(col_perf) + 1), col_perf.keys())

# Setting y-axis label
plt.xlabel('Model Index')

# Setting y-axis label
plt.ylabel('Accuracy over 20 epochs')

# Adding a title
plt.title('Accuracies for 10 Models')

# Displaying the plot
plt.show()
```
